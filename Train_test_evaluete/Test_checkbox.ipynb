```
#!pip install --upgrade git+https://github.com/broadinstitute/keras-resnet
import keras
import keras_resnet

# automatically reload modules when they have changed
%load_ext autoreload
%autoreload 2
#import keras
import keras

# import keras_retinanet
from keras_retinanet import models
from keras_retinanet import layers
from keras_retinanet.utils.image import read_image_bgr, preprocess_image, resize_image
from keras_retinanet.utils.visualization import draw_box, draw_caption
from keras_retinanet.utils.colors import label_color
from keras_retinanet.utils.gpu import setup_gpu

# import miscellaneous modules
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import time

# use this to change which GPU to use
gpu = 0

# set the modified tf session as backend in keras
setup_gpu(gpu)
```

```
# models can be downloaded here: https://github.com/fizyr/keras-retinanet/releases
model_path = "/content/gdrive/My Drive/RETINANET_files/Snapshot_weight_folder/resnet50_csv_1701885.h5"
# load retinanet model
model = models.load_model(model_path, backbone_name='resnet50') # https://www.tensorflow.org/api_docs/python/tf/keras/models/load_model

#def convert_model(model, nms=True, class_specific_filter=True, anchor_params=None):
# if the model is not converted to an inference model, use the line below
# see:model.predict_on_batch
model = models.convert_model(model)
#print(model.summary())

# load label to names mapping for visualization purposes
labels_to_names = {0: 'tick', 1: 'number',2: 'elipse'}
```

```
image = read_image_bgr('/content/gdrive/My Drive/RETINANET_files/images_4/323.jpg')

#copy to draw on
draw = image.copy()
draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB) # Change from BGR (OpenCV format) to RGB (Matlab format) to fit Matlab output

# preprocess image for network
image = preprocess_image(image)#Tensors in [0, 255]. This function maps pixel values to[-1, 1] and resizes to match the InceptionV1 network.

#image, scale = resize_image(image)# https://github.com/fizyr/keras-retinanet/blob/master/keras_retinanet/utils/image.py

# process image
start = time.time()
boxes, scores, labels = model.predict_on_batch(np.expand_dims(image, axis=0))

print("processing time: ", time.time() - start)
 
# visualize 
#print(labels,99909)
for box, score, label in zip(boxes[0], scores[0], labels[0]):
   
    # scores are sorted so we can break
    if score <0.5:
        break
     
    color = label_color(label)
    
    b = box.astype(int)
    draw_box(draw, b,color=color)
    
    caption = "{} {:.3f}".format(labels_to_names[label], score)
    draw_caption(draw, b, caption)
print(scores[0][0],boxes[0][0],labels[0][0])   
print(scores[0][1],boxes[0][1],labels[0][1]) 
print(scores[0][2],boxes[0][2],labels[0][2]) 
print(scores[0][3],boxes[0][3],labels[0][3]) 
plt.figure(figsize=(15, 15))
plt.axis('off')
plt.imshow(draw)
plt.show()
```
